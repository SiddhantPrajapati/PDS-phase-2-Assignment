```python
#Graph plot
import pandas as pd
import matplotlib.pyplot as plt
cal = []
far = []
def calcius(f):
  c = (f-32)*5/9
  cal.append(c)
for f in range(0,300):
  far.append(f)
  calcius(f)
plt.xlim(xmin=-18)
plt.ylim(ymin=0)
plt.xlim(xmax=150)
plt.ylim(ymax=300)
plt.plot(cal, far)
plt.show()
```

```python
#histogram
import numpy as np
import matplotlib.pyplot as plt
x = 20*np.random.randn(10000)
plt.hist(x,25,range=(-50,50),histtype='stepfilled',align='mid',label='Test Data')
plt.legend()
plt.title('Histogram')
plt.show()
```

```python
#pie chart
import matplotlib.pyplot as plot
val=[2,7,6,9,1,3]
colors=['b','g','r','c','m','y']
labels=['A','B','C','D','E','F']
explode=[0,0.1,0,0,0.2,0]
plt.pie(val,colors=colors,labels=labels,explode=explode,autopct='%1.1f%%',counterclock=False,shadow=True)
plt.title('Values')
plt.show()
```

```python
#bar charts
import matplotlib.pyplot as plot
val=[2,7,6,9,1,3]
colors=['b','g','r','c','m','y']
labels=['A','B','C','D','E','F']
widths=[0.4,0.5,0.6,0.7,0.5,0.7]
plt.bar(range(0,6),val,width=widths,color=colors,align='center')
plt.show()
```

```python
#Perform the pie plotting, bar chart, scatter plotting, histogram plotting for sample data.
#Pie plot graph
import matplotlib.pyplot as plt
import pandas as pd
count = []
data = pd.read_csv('/content/drive/MyDrive/Python_File/Flipkart Mobile - 2.csv')
model = data["brand"].tolist()
mset = set(model)
mlist = list(mset)
for i in mlist:
  count.append(model.count(i))
label = ['Apple','Poco','Realme','Samsung','Xiome']
plt.pie(count,labels=label,autopct='%1.1f%%',counterclock=False,shadow=True)
plt.show()
print('Count of Mobile Brand Available in Dataset')
```

```python
#gapminder data for specific any country and its population
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

mind=pd.read_csv('/content/drive/MyDrive/Python_File/Gap minder data.csv')

#convert csv into dataframe
x = [i*1 for i in range(1952,2010,5)]
country = list(set(mind["country"].tolist()))
population = mind["pop"].tolist()
#gdp = mind["gdpPercap"].tolist()
con = list(np.sort(country))

#to population data
pop = np.zeros(shape=(142,12))
i=0
for j in range(0,142):
  for k in range(0,12):
    pop[j][k] = round(population[i])
    i=i+1

ind_pop=pop[con.index('India')]
pak_pop=pop[con.index('Pakistan')]
cn_pop=pop[con.index('China')]
us_pop=pop[con.index('United States')]

plt.xlabel("Year")
plt.ylabel("Population")

#You can compare any country's population data
plt.plot(x,ind_pop,label='India',color='blue')
plt.plot(x,pak_pop,label="Pakistan",color='red')
plt.plot(x,cn_pop,label="China",color='yellow')
plt.plot(x,us_pop,label="U.S.",color='green')

plt.legend()
plt.show()
```

```python
#gapminder data for any specific country and gdp_per_capital
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

mind=pd.read_csv('/content/drive/MyDrive/Python_File/Gap minder data.csv')

#convert csv into dataframe

x = [i*1 for i in range(1952,2010,5)]
country = list(set(mind["country"].tolist()))
gdp = mind["gdpPercap"].tolist()
con = list(np.sort(country))

#to population data
GDP = np.zeros(shape=(142,12))
i=0
for j in range(0,142):
  for k in range(0,12):
    GDP[j][k] = round(gdp[i])
    i=i+1
  
ind_gdp=GDP[con.index('India')]
pak_gdp=GDP[con.index('Pakistan')]
cn_gdp=GDP[con.index('China')]
us_gdp=GDP[con.index('United States')]

plt.xlabel("Year")
plt.ylabel("GDP_Per_Capita")

#You can compare any country's gdp per capital data
plt.plot(x,ind_gdp,label='India',color='blue')
plt.plot(x,pak_gdp,label="Pakistan",color='red')
plt.plot(x,cn_gdp,label="China",color='yellow')
plt.plot(x,us_gdp,label="U.S.",color='green')

plt.legend()
plt.show()
```

```python
#gapminder data for any country and average life of people in country
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

mind=pd.read_csv('/content/drive/MyDrive/Python_File/Gap minder data.csv')

#convert csv into dataframe

x = [i*1 for i in range(1952,2010,5)]
country = list(set(mind["country"].tolist()))
life = mind["lifeExp"].tolist()
con = list(np.sort(country))

#to average life span data
Life = np.zeros(shape=(142,12))
i=0
for j in range(0,142):
  for k in range(0,12):
    Life[j][k] = round(life[i])
    i=i+1
  
ind_life=Life[con.index('India')]
pak_life=Life[con.index('Pakistan')]
cn_life=Life[con.index('China')]
us_life=Life[con.index('United States')]

plt.xlabel("Year")
plt.ylabel("Average Age")

#You can compare any country's population data
plt.plot(x,ind_life,label='India',color='blue')
plt.plot(x,pak_life,label="Pakistan",color='red')
plt.plot(x,cn_life,label="China",color='yellow')
plt.plot(x,us_life,label="U.S.",color='green')

plt.legend()
plt.show()
```
