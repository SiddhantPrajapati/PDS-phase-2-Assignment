```python
#Perform data plotting and visualisation in london city temp,rain and city data.
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
temp = pd.read_csv('/content/drive/MyDrive/Python_File/London_tempreture.csv')

#Convert it into dataframe
df = pd.DataFrame(temp)

#Remove column 1
df.drop(df.columns[0] , axis = 1, inplace = True)
mon = df['Month'].tolist()
tmax = df['Tmax'].tolist()
tmin = df['Tmin'].tolist()
print(mon)
plt.plot(mon,tmax, label='Max Tempreture', color = 'red')
plt.plot(mon,tmin, label='Min Tempreture', color = 'blue')
plt.xlabel('Month')
plt.ylabel('Tempreture')
plt.legend()
plt.show()
```

```python
#Bar chart on rain
rain = df['Rain'].tolist()
sun = df['Sun'].tolist()
plt.xlabel('Month')
plt.ylabel('Average Rain')
plt.bar(mon,rain,label='Rain')

plt.legend()
plt.show()
```
